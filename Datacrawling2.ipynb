```python
import requests
from bs4 import BeautifulSoup
```

```python
def load(url):
    try:
        page=requests.get(url)
        if page.status_code==200:
            print('success')
            return page.text
        else:
            print(page.status_code)
    except Exception as e:
        print(e)
        
    
        
            
        
```

```python
url="https://bbdu.ac.in/"
data=load(url)
```

```python
soup=BeautifulSoup(data,'html')
```
